```python
p501 = pd.read_csv('/Users/DeepDhindsa/Desktop/COSC4426/PalmerPenguins.csv')
p501
```

```python
p501.drop(columns=['Comments', 'Stage'], inplace=True)
p501
```

```python
plt.figure(figsize=(15,12))
sea.countplot(x=p501['Species'], alpha=0.65, saturation=0.85)
plt.title('Species Distributions', size=13, y =1.01)
plt.xlabel("")
plt.tick_params(axis='both', labelsize=12)
```

```python
sea.set_theme(style="white")
g1=sea.relplot(data=p500, alpha=0.6,
              x='Culmen Length(mm)', y='Culmen Depth(mm)', hue="Island", size="Body Mass(g)", palette="Set1", sizes =(40,250), height =7)
g1.despine(left=True, bottom=True)
g1.set_axis_labels("Bill Length", "Bill Depth")
```

```python
ax1 = plt.figure(figsize= (13,12))
ax1= sea.boxplot(data =p500, x='Species', y='Body Mass (g)',palette ="Set2")

ax1.set_ylabel('Body Mass(g)')
ax1.set_xlabel("")
ax1.set_title('Body Mass by Species')
ax1.tick_params(axis='x', labelsize=12)
```

```python
g1=plt.figure(figsize=(16,16))
g1=sea.catplot(data=p500, kind="bar", x='Species', y='Body Mass (g)', hue="Sex", ci="sd", palette="dark", alpha=.6, height =10)
g1.despine(left=True)
g1.set_axis_labels("", "Body Mass (g)")
g1.legend.set_title("")
```
